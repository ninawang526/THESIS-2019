```python
#### showing overlap ###

def overlap(group1, group2):    
    group1_usage = defaultdict(lambda: 0)
    for word in group1:
        group1_usage[word] += 1

    group2_usage = defaultdict(lambda: 0)
    for word in group2:
        group2_usage[word] += 1
    
    overlaps = defaultdict(lambda: 0)
        
    group1_unique = defaultdict(lambda: 0)
    for word in group1_usage:
        diff = group1_usage[word] - group2_usage[word]
        group1_unique[word] = max(0, diff)
        if diff >= 0:
            overlaps[word] = group2_usage[word]
        
    group2_unique = defaultdict(lambda: 0)
    for word in group2_usage:
        diff = group2_usage[word] - group1_usage[word]
        group2_unique[word] = max(0, diff)
        if diff >= 0:
            overlaps[word] = group1_usage[word]
    
    # sanity check
    for word in group1:
        assert (overlaps[word] + group1_unique[word] == group1_usage[word])
    for word in group2:
        assert (overlaps[word] + group2_unique[word] == group2_usage[word])
    
    return group1_unique, group2_unique, overlaps


sample1 = left_collocations
sample2 = right_collocations

left_unique, right_unique, overlaps = overlap(sample1, sample2)
    
num_overlaps = sum(overlaps.values())
num_left_unique = sum(left_unique.values())
num_right_unique = sum(right_unique.values())


print ("{}, or {} out of {}, are overlaps in {}".format(float(num_overlaps)/len(sample1), num_overlaps, len(sample1), "left_collocations"))
print ("")
print ("{}, or {} out of {}, are overlaps in {}".format(float(num_overlaps)/len(sample2), num_overlaps, len(sample2), "right_collocations"))



sorted_left_unique = sorted([(k,v) for k,v in left_unique.items() if v>0],key=lambda x:x[1],reverse=True)
sorted_right_unique = sorted([(k,v) for k,v in right_unique.items() if v>0],key=lambda x:x[1],reverse=True)


plt.figure(figsize=(16,8))
plt.xticks(rotation=90)

plt.plot([tup[0] for tup in sorted_left_unique[:50]], [tup[1] for tup in sorted_left_unique[:50]])
# plt.plot(sorted_left_unique.keys()[:50], sorted_left_unique.values()[:50])

plt.legend(['left', 'right'], loc='upper left')
plt.show()

plt.figure(figsize=(16,8))
plt.xticks(rotation=90)
plt.plot([tup[0] for tup in sorted_right_unique[:50]], [tup[1] for tup in sorted_right_unique[:50]])
plt.show()

top_left = [tup[0] for tup in sorted_left_unique[:50]]
print ((" ").join(top_left))
top_right = [tup[0] for tup in sorted_right_unique[:50]]
print ((" ").join(top_right))
```

```python
import gensim
from nltk import tokenize

docs = []
for article in articles:
    sentences = tokenize.sent_tokenize(article.text)
    sentences = [gensim.utils.simple_preprocess(s) for s in sentences]
    docs += sentences

model = gensim.models.Word2Vec(
    docs,
    size=150,
    window=10,
    min_count=2,
    workers=10)

model.train(docs, total_examples=len(docs), epochs=10)
```
